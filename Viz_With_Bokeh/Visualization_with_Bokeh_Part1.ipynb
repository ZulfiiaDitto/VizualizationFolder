Bohek library combines the following: 
* Python - for content and functionalities.
* JavaScript (library BohekJS) - backend for translating your code to Javascript and HTML to display in browser.


Workflow 
1. Create a figure 
2. Define your data source
3. Choose the glyphs - kind, color, etc4.Define how to place it into figure

Glyphs are the basic visual building blocks of Bokeh plots. This includes elements such as lines, rectangles, squares, wedges, or the circles of a scatter plot (Bokeh Documenttation).

```python
# we need to wrangle population ds -> lets group by the region and take mean of population 
populationTotal = population.merge(regions[['Country', 'Group']], on = 'Country', how = 'left')
populationTransp = populationTotal.groupby(['Group'], as_index = False).mean()
populationLong = pd.melt(populationTransp, id_vars=['Group'], 
                         value_vars = populationTransp.columns[1:],
                         value_name = 'Population',
                         var_name = 'Year')
populationLong.head(2)
```

```python
life_exp = life_expectancy.merge(regions[['Country', 'Group']], on = 'Country', how = 'left')
aggr_life_exp = life_exp.groupby('Group').mean()
aggr_life_exp.head(2)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Legend, ColorBar
from bokeh.io import output_notebook
from bokeh.palettes import Set1_6

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()
```

```python
### simple graph without hovertool and columns source 

graphExpect = figure(
    width= 1200,
    height=300,
    x_axis_label='years',
    y_axis_label='population life expectation (in years)', 
    title = 'Averaged Grouped life expectation over the years',)

graphExpect.add_layout(Legend(), 'right')

for i in range(len(aggr_life_exp)):
  graphExpect.line(x=aggr_life_exp.columns, y = aggr_life_exp.iloc[i], color= Set1_6[i], 
                   legend_label = aggr_life_exp.index[i], line_width=2)
  graphExpect.circle(x=aggr_life_exp.columns, y = aggr_life_exp.iloc[i], color= Set1_6[i], 
                    line_width=0.5)

show(graphExpect)
```

```python
from ipywidgets import interact 
from math import pi
from bokeh.models import RangeSlider, HoverTool
from bokeh.layouts import layout
```

The ColumnDataSource (CDS) is the core of most Bokeh plots. It provides the data to the glyphs of your plot.

When you pass sequences like Python lists or NumPy arrays to a Bokeh renderer, Bokeh automatically creates a ColumnDataSource with this data for you. However, creating a ColumnDataSource yourself gives you access to more advanced options. - Bohek documentation

```python
### graph with source and hovertool
# utilizing columns source gave a lot of advantages -> hovertool 

regionEurope = populationLong.loc[populationLong.Group.str.startswith('Europe')].copy()

popSourceEuro = ColumnDataSource(regionEurope)

hover = [('Year', '@Year'), ('Population', '@Population')]    

graphPopEuro = figure(width= 1100,
    height=300, x_range = regionEurope.Year.unique(),
    x_axis_label='years',
    y_axis_label='Average of Population', 
    title = 'Averaged Population over the years in America region', tooltips=hover)

graphPopEuro.vbar(x='Year', top='Population', width=0.9, source=popSourceEuro, line_color = 'black', color = '#9c9ede')

graphPopEuro.xaxis.major_label_orientation = pi/2

show(graphPopEuro)
```

```python
# Add range slider for years 
regionAmerica = populationLong.loc[populationLong.Group == "America"].sort_values(by = 'Year')
regionAmerica['Year'] = regionAmerica['Year'].astype(int)

popSourceAmerica = ColumnDataSource(regionAmerica)

hover = [('Year', '@Year'), ('Population', '@Population')]    

graphPopAmerica = figure(width= 1100,
    height=300, x_range = (regionAmerica.Year.min()-1, regionAmerica.Year.max()+1),
    x_axis_label='years',
    y_axis_label='Average of Population', 
    title = 'Averaged Population over the years in America region', tooltips=hover)

graphPopAmerica.vbar(x='Year', top='Population', width=0.9, source=popSourceAmerica, line_color = 'black', color = '#9c9ede')

graphPopAmerica.xaxis.major_label_orientation = pi/2

#show(graphPopAmerica)

range_slider = RangeSlider(
    title="Adjust year range", 
    start=regionAmerica.Year.min(),  # set the minimum year
    end=regionAmerica.Year.max(),  # set the maximum year
    step=1,  # increments for the slider
    value=(graphPopAmerica.x_range.start, graphPopAmerica.x_range.end),  
    )
range_slider.js_link("value", graphPopAmerica.x_range, "start", attr_selector=0)
range_slider.js_link("value", graphPopAmerica.x_range, "end", attr_selector=1)

withRange = bokeh.layouts.layout([[range_slider], [graphPopAmerica]])

show(withRange)
```

```python
#### use of scatterplot and colormap 
from bokeh.transform import linear_cmap
from bokeh.models.annotations import Tooltip
from bokeh.layouts import column
from bokeh.plotting import figure, show
from bokeh.transform import jitter

AmericaRegionLong = pd.melt(populationTotal.loc[populationTotal.Group == 'America'], id_vars=['Country'], 
                         value_vars = populationTotal.columns[1:-1],
                         value_name = 'Population',
                         var_name = 'Year')

AmericaRegionLong.Year = AmericaRegionLong.Year.astype(int)

cmap = linear_cmap(field_name='Population', palette="Spectral6", low=AmericaRegionLong.Population.min(), high=AmericaRegionLong.Population.max())

source = ColumnDataSource(AmericaRegionLong)
years = sorted(AmericaRegionLong.Year.unique())
hover = [('Country', '@Country'), ('Population', '@Population'), ('Year', '@Year')]

plot = figure(width=1200, height=300, title="Population distribution by country and year in America Region",
              tooltips = hover)
plot.xaxis.ticker = years
plot.xaxis.major_label_orientation = pi/2

plot.scatter(x=jitter('Year', 0.4), y='Population', #marker = 'diamond',
             size=9, alpha=0.4, source=AmericaRegionLong, color = cmap)

color_bar = ColorBar(color_mapper=cmap['transform'], width=10)
plot.add_layout(color_bar, 'right')
show(plot)
```

```python
####### Bonus - ineraction with Ipywidgets ################

### adding the interactivity with ipywidgets 
from bokeh.io.notebook import push_notebook

life_exp_melt = pd.melt(aggr_life_exp.reset_index(),id_vars=['Group'], 
                         value_vars = populationTotal.columns[1:-1],
                         value_name = 'Population',
                         var_name = 'Year')

def update(region):
  newRegion = life_exp_melt.loc[life_exp_melt.Group == region]
  line.data_source.data.update(newRegion)
  circle.data_source.data.update(newRegion)
  push_notebook()


region_default = 'America'
exp_default = life_exp_melt.loc[life_exp_melt.Group==region_default]

sourceRegion = ColumnDataSource(exp_default)

graphExpect = figure(
    width= 1200,
    height=300,
    x_axis_label='years',
    y_axis_label='population life expectation (in years)', 
    title = 'Averaged Grouped life expectation over the years',)

line = graphExpect.line(x='Year', y = 'Population', source = sourceRegion,
                    line_width=2,)
circle = graphExpect.circle(x='Year', y = 'Population', source = sourceRegion,
                    line_width=0.5,)


show(graphExpect, notebook_handle=True)

interact(update, region = life_exp_melt.Group.unique())
```
